```python
import nltk
from nltk.parse import ShiftReduceParser
from nltk import CFG
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det NN
VP -> V NP
Det -> 'the'
NN -> 'boy' | 'ball'
V -> 'hit'
""")
p= ShiftReduceParser(grammar)
sentence = "the boy hit the ball".split()

print("STACK\t\tINPUT \t\tACTION")
stack=[]
input=sentence[:]

for w in sentence:
    stack.append(w)
    input.pop(0)
    print(stack,"\t",input,"\tSHIFT")
print(stack,"\t",input,"\tREDUCE\n")

for tree in p.parse(sentence):
    tree.pretty_print()
    
```
